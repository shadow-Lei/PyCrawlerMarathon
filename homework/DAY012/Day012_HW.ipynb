### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup=BeautifulSoup(r.text, "html")
for s in soup.find(class_='part_list_2').findChildren('h3'):
    print(s.find(class_='date').text,s.find('em').text,s.find('a').text)
    
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python

news={}
for s in soup.find(class_='part_list_2').findChildren('h3'):
    datetime=s.find(class_='date').text
    tag=s.find('em').text
    title=s.find('a').text
    news.setdefault(tag,[])
    news[tag].append(datetime+'  '+title.replace('\u3000',' '))

for k in news.items():
    tag=k[0]
    print(tag+':')
    for title in k[1]:
        print(title)
```

### ③ 哪一個類別的文章最多

```python
datacount=[]
for k in news.items():
    tag=k[0]
    count=len(k[1])
    print(tag,count,sep='\t')
    datacount.append([tag,count])
```

```python
import pandas as pd
df = pd.DataFrame(datacount,columns=['tag','count'])
df
```
